```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.inews.id/multimedia/infografis'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
list_content = soup.select ('.multimedia')
```

```python
list_content
```

```python
list_content = soup.select ('article')
```

```python
list_content
```

```python
len(list_content)
```

```python
results = []
for item in list_content:
  data = dict()

  data['title'] = item.select_one('.cardTitle').text
  data['url'] = item.select_one('.cardArticle a').get('href')
  data['url image'] = item.select_one('.thumbCard').get('src')
  results.append(data)
```

```python
pd.DataFrame(results)
```
